## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
numQuestoesDic = {'Matemática': 24,
             'Português': 18,
             'Geografia': 8,
             'Inglês': 8,
             'História': 8,
             'Física': 12,
             'Química': 12}
````

````python
# 1) Seu código aqui

#percentual de acertos de Matemática:
nquestoesmat = numQuestoesDic['Matemática']
df_mat_perc = (df_mat*100) / nquestoesmat
df_mat_perc
````

````python
#percentual de acertos de Português:
nquestoespor = numQuestoesDic['Português']
df_por_perc = (df_por*100) / nquestoespor
df_por_perc
````

````python
#percentual de acertos de Geografia:
nquestoesgeo = numQuestoesDic['Geografia']
df_geo_perc = (df_geo*100) / nquestoesgeo
df_geo_perc
````

````python
#percentual de acertos de inglês:
nquestoesing = numQuestoesDic['Inglês']
df_ing_perc = (df_ing*100) / nquestoesing
df_ing_perc
````

````python
#percentual de acertos de História:
nquestoeshis = numQuestoesDic['História']
df_his_perc = (df_his*100) / nquestoeshis
df_his_perc
````

````python
#percentual de acertos de Física:
nquestoesfis = numQuestoesDic['Física']
df_fis_perc = (df_fis*100) / nquestoesfis
df_fis_perc
````

````python
#percentual de acertos de Química:
nquestoesqui = numQuestoesDic['Química']
df_qui_perc = (df_qui*100) / nquestoesqui
df_qui_perc
````

````python
# 2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  

total_de_acertos = df_mat + df_por + df_geo + df_his + df_ing + df_fis + df_qui
total_de_acertos
````

````python
# 3. Calcule o porcentual geral de cada aluno. 

total_de_acertos/90 * 100
````

````python
# 4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  
total_gt_45 = total_de_acertos[total_de_acertos['Qt_acertos'] > 45]
total_gt_45.count()

aprovados = True
n_aprovados = total_de_acertos > 45
n_aprovados.value_counts()
````
